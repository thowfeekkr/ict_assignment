```python

expenses = []


while True:
    date = input("Enter date (YYYY-MM-DD): ")
    category = input("Enter category: ")
    amount = float(input("Enter amount: "))

    expenses.append([date, category, amount])

    choice = input("Add another expense? (yes/no): ")
    if choice == "no":
        break

total_expense = 0

for exp in expenses:
    total_expense = total_expense + exp[2]

print("\nTotal Monthly Expense:", total_expense)

categories = []
category_totals = []

for exp in expenses:
    cat = exp[1]
    amt = exp[2]

    if cat in categories:
        index = categories.index(cat)
        category_totals[index] = category_totals[index] + amt
    else:
        categories.append(cat)
        category_totals.append(amt)

print("\nCategory-wise Spending:")
for i in range(len(categories)):
    print(categories[i], ":", category_totals[i])


budget = float(input("\nEnter your monthly budget: "))

if total_expense > budget:
    print(" Budget Breached!")
else:
    print("You are within the budget")


dates = []
date_totals = []

for exp in expenses:
    d = exp[0]
    amt = exp[2]

    if d in dates:
        index = dates.index(d)
        date_totals[index] = date_totals[index] + amt
    else:
        dates.append(d)
        date_totals.append(amt)

max_amount = date_totals[0]
max_date = dates[0]

for i in range(len(date_totals)):
    if date_totals[i] > max_amount:
        max_amount = date_totals[i]
        max_date = dates[i]

print("\nHighest Spending Day:", max_date)
print("Amount Spent:", max_amount)


max_category = categories[0]
max_spend = category_totals[0]

for i in range(len(category_totals)):
    if category_totals[i] > max_spend:
        max_spend = category_totals[i]
        max_category = categories[i]

print("\nDominant Spending Category:", max_category)
```
